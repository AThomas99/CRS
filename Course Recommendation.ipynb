```python
import numpy as np
import pandas as pd
```

```python
course_df = pd.read_csv('course.csv')
ratings_df = pd.read_csv('ratings.csv')
```

```python
course_df.head()
```

```python
course_df['title']
```

```python
course_df['interest'] = course_df.interest.str.split('|')
```

```python
course_df.head()
```

```python
courseWithInterest_df = course_df.copy()

for index, row in course_df.iterrows():
    for interest in row['interest']:
        courseWithInterest_df.at[index, interest] = 1

courseWithInterest_df = courseWithInterest_df.fillna(0)
courseWithInterest_df.head()
```

```python
ratings_df.head()
```

```python
ratings_df = ratings_df.drop('timestamp', 1)
ratings_df.head()
```

```python
userInput = [
            {'title':'Ordinary Diploma in Computer Engineering', 'rating':5},
            {'title':'Ordinary Diploma Multimedia and Film Technology', 'rating':3.5},
            {'title':'Ordinary Diploma in Food Science and Technology', 'rating':2},
            {'title':'Ordinary Diploma in Information Technology', 'rating':5},
            {'title':'Ordinary Diploma in Electronics and Telecommunications Engineering', 'rating':4.5}
         ]
inputCourse = pd.DataFrame(userInput)
inputCourse
```

```python
#Filtering out the movies by title
inputId = course_df[course_df['title'].isin(inputCourse['title'].tolist())]

#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputCourse = pd.merge(inputId, inputCourse)

#Dropping information we won't use from the input dataframe
inputCourse = inputCourse.drop('interest', 1)

inputCourse
```

```python
#Filtering out the courses from the input
userCourses = courseWithInterest_df[courseWithInterest_df['courseId'].isin(inputCourse['courseId'].tolist())]
userCourses
```

```python
#Resetting the index to avoid future issues
userCourses = userCourses.reset_index(drop=True)

#Dropping unnecessary issues due to save memory and to avoid issues
userInterestTable = userCourses.drop('courseId', 1).drop('title', 1).drop('interest', 1)
userInterestTable
```

```python
inputCourse['rating']
```

```python
userProfile = userInterestTable.transpose().dot(inputCourse['rating'])
userProfile
```

```python
#Now let's get the genres of every movie in our original dataframe
interestTable = courseWithInterest_df.set_index(courseWithInterest_df['courseId'])

#And drop the unnecessary information
interestTable = interestTable.drop('courseId', 1).drop('title', 1).drop('interest', 1)
interestTable.head()
```

```python
interestTable.shape
```

```python
#Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((interestTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
#Sort our recommendations in descending order
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

#Just a peek at the values
recommendationTable_df.head()
```
